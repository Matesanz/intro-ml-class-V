Titanic Spaceship Kaggle Competition
========================

Lets dig into the details of the Titanic Spacship dataset and explore the features that will help us predict survival rates.

```python
import pandas as pd


DATA_FOLDER = "../data/spaceship-titanic/"
TRAIN_FILE_PATH = DATA_FOLDER + "train.csv"
TEST_FILE_PATH = DATA_FOLDER + "test.csv"
SAMPLE_SUBMISSION_FILE_PATH = DATA_FOLDER + "sample_submission.csv"


train_df = pd.read_csv(TRAIN_FILE_PATH)
test_df = pd.read_csv(TEST_FILE_PATH)

train_df.head(10)
```

# Preprocessing

```python
clean_train_df = train_df.dropna()
```

```python
clean_train_df.head(2)
```

# Train Test Split

```python
from random import seed
from sklearn.model_selection import train_test_split

TEST_SIZE = 0.2
SEED = 42

X = clean_train_df.select_dtypes(include="number")
y = clean_train_df["Transported"]


X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=TEST_SIZE, random_state=SEED)
```

# Train the Model

```python
from sklearn.tree import DecisionTreeClassifier


model = DecisionTreeClassifier(random_state=SEED)
model.fit(X_train, y_train)
```

# Evaluate the model

```python
model.score(X_val, y_val)
```

```python
from sklearn.metrics import accuracy_score


y_pred = model.predict(X_val)
acc = accuracy_score(y_val, y_pred)
print(f"Accuracy: {acc:.3f}")
```

# Send data to Kaggle

```python
test_df.head()
```

```python
X_test = test_df.select_dtypes(include="number")
X_test.fillna(0, inplace=True)
y_pred = model.predict(X_test)
y_pred
```

```python
submission_df = pd.DataFrame({"PassengerId": test_df["PassengerId"], "Transported": y_pred,})
submission_df.head(2)
```

```python
submission_file_path = DATA_FOLDER + "submission.csv"
submission_df.to_csv(submission_file_path, index=False)
```
